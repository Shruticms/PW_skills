```python
#Implement a generator function that reads a file line by line and yields each line as a string.
def read_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
```

1.	What is the difference between a function and a method in Python?
- In Python, both functions and methods are blocks of reusable code designed to perform specific tasks, but their key difference lies in their association with classes and objects.
Function:
A function is a standalone block of code that can be defined independently and called from anywhere in the program. It operates on data explicitly passed to it as arguments and can optionally return a value. Functions are used for general-purpose tasks that are not inherently tied to a specific object's state or behavior.

A method is a function that is associated with a class or an object (an instance of a class). It is defined within a class and operates on the data (attributes) of that specific object. Methods are called on instances of a class and implicitly receive the object itself as their first argument (conventionally named self). Methods are used to encapsulate behavior related to an object's state and actions.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- The map(), filter(), and reduce() functions in Python are higher-order functions used for processing iterables, each serving a distinct purpose:
•	map(function, iterable, ...):
o	Purpose: Applies a given function to each item in one or more iterables and returns an iterator yielding the results.
o	Output: An iterator (which can be converted to a list, tuple, etc.) containing the transformed elements, with the same number of elements as the input iterable(s).
o	Use Case: When you need to transform each element of a sequence in a uniform way (e.g., doubling numbers, converting strings to uppercase).
Python
    numbers = [1, 2, 3, 4]
    squared_numbers = list(map(lambda x: x * x, numbers))
    # squared_numbers will be [1, 4, 9, 16]
•	filter(function, iterable):
o	Purpose: Constructs an iterator from elements of an iterable for which the function returns True.
o	Output: An iterator (which can be converted to a list, tuple, etc.) containing only the elements that satisfy the condition defined by the function. The number of elements in the output can be less than or equal to the input.
o	Use Case: When you need to select specific elements from a sequence based on a condition (e.g., filtering out even numbers, selecting positive values).
Python
    numbers = [1, 2, 3, 4, 5, 6]
    even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
    # even_numbers will be [2, 4, 6]
•	reduce(function, iterable[, initial]):
o	Purpose: Applies a function of two arguments cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. (Note: reduce is in the functools module in Python 3, not a built-in function).
o	Output: A single value that is the result of the cumulative application of the function.
o	Use Case: When you need to aggregate or combine all elements of a sequence into a single result (e.g., summing all elements, finding the product, finding the maximum).
Python
    from functools import reduce
    numbers = [1, 2, 3, 4]
    sum_of_numbers = reduce(lambda x, y: x + y, numbers)
    # sum_of_numbers will be 10
